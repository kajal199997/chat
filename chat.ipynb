```python
from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer
import os
```

```python
kgp = ChatBot('Bot')
```

```python
kgp.set_trainer(ListTrainer)
```

```python
for f in os.listdir('chatterbot-corpus-master\\chatterbot-corpus-master\\chatterbot_corpus\\data\\english') :
    data = open('chatterbot-corpus-master\\chatterbot-corpus-master\\chatterbot_corpus\\data\\english\\' + f).readlines()
    kgp.train(data)   
```
